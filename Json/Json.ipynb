```python
import pandas as pd
import json
import random
```

```python
obd = pd.read_excel('Julio_BD.xlsx')
```

```python
ebd = ['TO_Semana_1','TO_Semana_2','TO_Semana_3','TO_Semana_4','TO_Semana_5','TO_Semana_6','FONO_Semana_1','FONO_Semana_2','FONO_Semana_3','FONO_Semana_4','FONO_Semana_5','FONO_Semana_6','PSICO_Semana_1','PSICO_Semana_2','PSICO_Semana_3','PSICO_Semana_4','PSICO_Semana_5','PSICO_Semana_6','FISIO_Semana_1','FISIO_Semana_2','FISIO_Semana_3','FISIO_Semana_4','FISIO_Semana_5','FISIO_Semana_6','RHC_Semana_1','RHC_Semana_2','RHC_Semana_3','RHC_Semana_4','RHC_Semana_5','RHC_Semana_6','CLI_Semana_1','CLI_Semana_2','CLI_Semana_3','CLI_Semana_4','CLI_Semana_5','CLI_Semana_6','TGR_Semana_1','TGR_Semana_2','TGR_Semana_3','TGR_Semana_4','TGR_Semana_5','TGR_Semana_6']

obd = obd.drop(columns=ebd)
```

```python
nuevas_filas = []

# Función para generar una factura consecutiva
def generar_factura_consecutiva(inicio, indice):
    return f"FE{inicio + indice}"

# Definir las columnas de terapias
terapias = ['Suma T.O', 'Suma FONO', 'Suma Psico', 'Suma Fisio', 'Suma RHC', 'Suma CLI', 'Suma TGR']

# Iterar sobre cada fila del DataFrame original
for index, row in obd.iterrows():
    for terapia in terapias:
        # Repetir la fila según el valor de la columna de terapia
        for _ in range(row[terapia]):
            nueva_fila = row.copy()
            nueva_fila['Terapia'] = terapia.replace('Suma ', '')  # Remover 'Suma ' del nombre de la terapia
            nuevas_filas.append(nueva_fila)

# Crear el nuevo DataFrame
df_nuevo = pd.DataFrame(nuevas_filas)

# Seleccionar solo las columnas necesarias
df_nuevo = df_nuevo[['ID','Nombre', 'Tipo de Documento', 'Terapia']]
df_nuevo = df_nuevo.rename(columns={'ID': 'numDocumentoIdentificacion'})
df_nuevo = df_nuevo.rename(columns={'Tipo de Documento': 'tipoDocumentoIdentificacion'})
df_nuevo['numDocumentoIdObligado"'] = "901183045"

# Obtener tipos de documentos únicos y asignarles una factura consecutiva
documentos_unicos = df_nuevo['numDocumentoIdentificacion'].unique()
factura_asignada = {}
inicio_factura = 2001  # Puedes cambiar este valor si necesitas empezar en otro número

for i, doc_id in enumerate(documentos_unicos):
    factura_asignada[doc_id] = generar_factura_consecutiva(inicio_factura, i)

df_nuevo['numFactura'] = df_nuevo['numDocumentoIdentificacion'].map(factura_asignada)
df_nuevo['numNota'] = None
df_nuevo['tipoNota'] = None
df_nuevo['codMunicipioResidencia'] = None
df_nuevo['codPaisOrigen'] = None
df_nuevo['codPaisResidencia'] = None
df_nuevo['codSexo'] = None #BDITI
df_nuevo['consecutivo'] = None
df_nuevo['fechaNacimiento'] = None #BDITI
df_nuevo['incapacidad'] = None 
df_nuevo['tipoUsuario'] = None 
df_nuevo['codPrestador'] = None 
df_nuevo['fechaInicioAtencion'] = None 
df_nuevo['numAutorizacion'] = None 
df_nuevo['codConsulta'] = None 
df_nuevo['modalidadGrupoServicioTecSal'] = None 
df_nuevo['grupoServicios'] = None 
df_nuevo['codServicio'] = None 
df_nuevo['finalidadTecnologiaSalud'] = None 
df_nuevo['causaMotivoAtencion'] = None 
df_nuevo['codDiagnosticoPrincipal'] = None 
df_nuevo['codDiagnosticoRelacionado1'] = None 
df_nuevo['codDiagnosticoRelacionado2'] = None 
df_nuevo['codDiagnosticoRelacionado3'] = None 
df_nuevo['tipoDiagnosticoPrincipal'] = None 
df_nuevo['vrServicio'] = None 
df_nuevo['conceptoRecaudo'] = None 
df_nuevo['valorPagoModerador'] = None 
df_nuevo['numFEVPagoModerador'] = None 
df_nuevo['consecutivo'] = None 

mapeo_servicios = {
    'T.O':'TERAPIA OCUPACIONAL INTEGRAL',
    'FONO':'TERAPIA FONOAUDIOLÓGICA INTEGRAL',
    'Psico':'TERAPIA PSICOLOGIA INTEGRAL',
    'Fisio':'TERAPIA FISICA INTEGRAL',
    'RHC':' TERAPIA FONOAUDIOLÓGICA INTEGRAL',
    'CLI':' TERAPIA FONOAUDIOLÓGICA INTEGRAL',
    'TGR':' TERAPIA FONOAUDIOLÓGICA INTEGRAL'
}
df_nuevo['Terapia'] = df_nuevo['Terapia'].map(mapeo_servicios)

df_nuevo.to_excel('validar.xlsx',index=False)
```

```python
# Crear el dataset con valores vacíos
bdjson ={
"numDocumentoIdObligado": None,
"numFactura": None,
"numNota": None,
"tipoNota": None,
"usuarios": [
        {
        "codMunicipioResidencia": None,
        "codPaisOrigen": None,
        "codPaisResidencia": None,
        "codSexo": None,
        "codZonaTerritorialResidencia": None,
        "consecutivo": None,
        "fechaNacimiento": None,
        "incapacidad": None,
        "numDocumentoIdentificacion": None,
        "tipoDocumentoIdentificacion": None,
        "tipoUsuario": None,
        "servicios": {
            "consultas": [
                    {
                    "codPrestador": None,
                    "fechaInicioAtencion": None,
                    "numAutorizacion": None,
                    "codConsulta": None,
                    "modalidadGrupoServicioTecSal": None,
                    "grupoServicios": None,
                    "codServicio": None,
                    "finalidadTecnologiaSalud": None,
                    "causaMotivoAtencion": None,
                    "codDiagnosticoPrincipal": None,
                    "codDiagnosticoRelacionado1": None,
                    "codDiagnosticoRelacionado2": None,
                    "codDiagnosticoRelacionado3": None,
                    "tipoDiagnosticoPrincipal": None,
                    "tipoDocumentoIdentificacion": None,
                    "numDocumentoIdentificacion": None,
                    "vrServicio": None,
                    "conceptoRecaudo": None,
                    "valorPagoModerador": None,
                    "numFEVPagoModerador": None,
                    "consecutivo": None
                    }
                ]
            }
        }
    ]
}
```

```python
# List to keep track of files that could not be created
failed_files = []

# Generar archivos JSON por cada paciente
for _, row in df_nuevo.iterrows():
    paciente_json = bdjson.copy()
    paciente_json['numDocumentoIdObligado'] = row.get('numDocumentoIdObligado', "901183045")
    paciente_json['numFactura'] = row.get('numFactura')
    paciente_json['numNota'] = row.get('numNota')
    paciente_json['tipoNota'] = row.get('tipoNota')
    
    # Rellenar la información del usuario
    usuario_info = paciente_json['usuarios'][0]
    usuario_info['codMunicipioResidencia'] = row.get('codMunicipioResidencia')
    usuario_info['codPaisOrigen'] = row.get('codPaisOrigen')
    usuario_info['codPaisResidencia'] = row.get('codPaisResidencia')
    usuario_info['codSexo'] = row.get('codSexo')
    usuario_info['consecutivo'] = row.get('consecutivo')
    usuario_info['fechaNacimiento'] = row.get('fechaNacimiento')
    usuario_info['incapacidad'] = row.get('incapacidad')
    usuario_info['numDocumentoIdentificacion'] = row.get('numDocumentoIdentificacion')
    usuario_info['tipoDocumentoIdentificacion'] = row.get('tipoDocumentoIdentificacion')
    usuario_info['tipoUsuario'] = row.get('tipoUsuario') 
    
    # Rellenar la información del servicio
    if usuario_info['servicios']:
        servicio_info = usuario_info['servicios']
        if servicio_info['consultas']:
            consulta_info = servicio_info['consultas'][0]
            consulta_info['numFEVPagoModerador'] = row.get('numFactura')
    else:
        failed_files.append(f"Faltan servicios para el paciente con documento {row.get('numDocumentoIdentificacion')}")
        
    # Guardar el JSON en un archivo
    archivo_json = f"paciente_{row['numDocumentoIdentificacion']}.json"
    try:
        with open(archivo_json, 'w') as f:
          json.dump(paciente_json, f, indent=4)
    except Exception as e:
        failed_files.append((archivo_json, str(e)))



# Informar sobre los archivos que no se pudieron crear
if failed_files:
    print("Los siguientes archivos no se pudieron crear:")
    for archivo, error in failed_files:
        print(f"Archivo: {archivo} - Error: {error}")
else:
    print("Todos los archivos se crearon exitosamente.")
```
